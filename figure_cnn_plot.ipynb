```
#plot accuracy
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import numpy as np

! mkdir plots

col = ['red','green','blue']
style = ['-','-','-'] #['-','--',':']

for n in range(1,7):

  fig = pyl.figure(figsize=(8,5))

  for k in range(0,3):
    #print(size)
    line, = pyl.plot(np.divide(size[n],size[n][0]/100.), acc[n][k], style[k], color=col[k])
    pyl.plot(np.divide(size[n],size[n][0]/100.), acc[n][k],"o", color=col[k])
     
    label = "Scaling factor = "+str(extrap[n][k])
    line.set_label(label)

  pyl.legend() 
  pyl.title(title[n])
  pyl.xlabel("IMAGE FILE SIZE [% OF ORIGINAL SIZE]")
  pyl.ylabel("CNN ACCURACY")
  

  auxName = "plots/acc_"+str(n)+".png"
  pyl.savefig(auxName)

  pyl.show()

! zip -r plots.zip plots
```
